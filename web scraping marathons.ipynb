```python
df_all['finishtime_dt'] = pd.to_datetime(df_all['FinishTimeConverted'])
```

```python
df_all['hours'] = pd.to_timedelta(df_all.FinishTimeConverted).apply(lambda x:x.seconds/3600)
```

```python
fig, ax = plt.subplots(figsize=(16,7.5))
sns.set_style("dark")
sns.violinplot(y = df_all.hours, x = df_all.SubeventName)
plt.title('Distribution of Marathon Times by Year')
plt.xlabel('');
```
